```
dataset = pd.read_csv('AAPL Mega.csv')
```

```
dataset
```

```
dataset_total = pd.concat([dataset, Sentiment_dataset], axis=1)
```

```
dataset_total = dataset_total.drop(dataset_total.index[-1])
```

```
dataset_total
```

```
dataset_total['percentage change'] = 100*(dataset_total['Close']-dataset_total['Open'])/dataset_total['Open']
dataset_total.loc[dataset_total['percentage change']<0,'Invest'] = int(0)
dataset_total.loc[dataset_total['percentage change']>=0, 'Invest'] = int(1)
dataset_total = dataset_total.drop(['Local time', 'Date', 'Close', 'percentage change'], 1)
```

```
dataset_total = dataset_total.dropna()
```

```
dataset_total
```

```
dataset_total = dataset_total.reset_index().drop(['index'],1)
```

```
dataset_total
```

```
X_train = []
y_train = []
training_set = dataset_total
for i in range(500, len(training_set)):
  X_train.append(training_set.iloc[i-500:i,:-1].values)
  y_train.append(training_set.iloc[i, -1])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```
X_train.shape
```

```
shape0 = X_train.shape[0]
shape1 = X_train.shape[1]
X_train = X_train.reshape(X_train.shape[0]*X_train.shape[1], 10)
X_train.shape
y_train = y_train.reshape(-1, 1)
```

```
X_train.shape
```

```
y_train.shape
```

```
from sklearn.preprocessing import MinMaxScaler
sc_x = MinMaxScaler(feature_range = (0,1))
X_train = sc_x.fit_transform(X_train)
```

```
X_train
```

```
y_train
```

```
X_train = np.reshape(X_train, (shape0, shape1, 10))
```

```
with tf.device('/gpu:0'):
  classifier = tf.keras.models.Sequential([
               tf.keras.layers.LSTM(units = 100, return_sequences = True, input_shape = (X_train.shape[1],10)),
                                    tf.keras.layers.Dropout(rate = 0.2),
                                    tf.keras.layers.LSTM(units = 100, return_sequences = True),
                                    tf.keras.layers.Dropout(rate = 0.2),
                                    tf.keras.layers.LSTM(units = 100, return_sequences = True),
                                    tf.keras.layers.Dropout(rate = 0.2),
                                    tf.keras.layers.LSTM(units = 100, return_sequences = True),
                                    tf.keras.layers.Dropout(rate = 0.2),
                                    tf.keras.layers.Dense(units = 1)])
```

```
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
classifier.fit(X_train, y_train, batch_size = 32, epochs = 10)
```
