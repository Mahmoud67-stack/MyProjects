#Importing the Important Libraries

```
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import math
```

## Importing the dataset

```
dataset = pd.read_csv('AMZN.USUSD_Candlestick_1_M_BID_01.03.2020-10.07.2020.csv')
#dataset['Local time'] = datetime.strptime(dataset['Local time'], full)
dataset.index = dataset['Local time']
dataset = dataset.drop('Local time', axis =1 )
dataset_original = dataset
dataset_original
```

```
dataset['percentage change'] = 100*(dataset['Close']-dataset['Open'])/dataset['Open']
```

```
dataset
```

```
dataset
```

```
dataset['percentage_pred'] = dataset['percentage change'].shift(-390)
```

```
dataset
```

```
dataset.loc[dataset['percentage_pred']<0,'Invest'] = int(0)
dataset.loc[dataset['percentage_pred']>=0, 'Invest'] = int(1)
dataset
```

```
X = dataset[['Open','High','Low','Close','Volume','percentage change']].values
y = dataset[['Invest']].values
print(X)
print(y)
```

```
X = X[:-390]
print(X)
y = y[:-390]
print(y)
```

```
X_train = X[:len(X)-390]
X_test = X[len(X)-390:]
y_train = y[:len(y)-390]
y_test = y[len(y)-390:]
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components = 3, kernel = 'rbf', random_state=0)
X_train = kpca.fit_transform(X_train)
X_test = kpca.transform(X_test)
```

```
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

```
print(y_test.shape)
y_pred = classifier.predict(X_test)
```

```
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
