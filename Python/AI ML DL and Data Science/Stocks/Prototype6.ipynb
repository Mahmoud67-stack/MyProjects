```
import pandas as pd
import numpy as np
import math
```

```
dataset = pd.read_csv('AMZN.USUSD_Candlestick_1_M_BID_01.03.2020-10.07.2020.csv')
dataset = dataset.drop('Local time', axis =1 )
```

```
dataset
```

```
X = dataset[['Close', 'Open', 'High', 'Low', 'Volume']].values
y = dataset['Close'].shift(-390)
```

```
X = X[:-390]
print(X)
y = y[:-390]
print(y)
```

```
X_train = X[:len(X)-390]
X_test = X[len(X)-390:]
y_train = y[:len(y)-390]
y_test = y[len(y)-390:]
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
lin_regressor.fit(X_train,y_train) # Here we are training the linearregression object on the data that we have
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2) # Here we are making an instance of the PolynomialFeatures class that helps as to make a matrix  with that form of X^0+X^1+X^2 for all the X data 
X_poly = poly_reg.fit_transform(X_train) # Here we are fitting the polynomial feature and transofrmin the result on the X dataset
lin_reg_2 = LinearRegression() # Here we make an instance of the LinearRegression class that helps us to claculate the linear coeficients of the equation b0*X^0+b1*X^1+b2*X^2 
lin_reg_2.fit(X_poly, y_train) # Here we are fitting the datasets that we have from the polynomilafeature class and the y dataset in other words making the linear regression for the data
```

```
 y_pred = lin_reg_2.predict(poly_reg.transform(X_test))
```

```
import sklearn.metrics as sm
print("R2 score =", round(sm.r2_score(y_test, y_pred)))
```
