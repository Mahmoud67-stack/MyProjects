# Artificial Neural Network

### Importing the libraries

```
import pandas as pd
import numpy as np
import tensorflow as tf
```

```
tf.__version__
```

## Part 1 - Data Preprocessing

### Importing the dataset

```
dataset = pd.read_csv('Modified.csv')
dataset
```

```
dataset['pred_close'] = dataset['Close'].shift(-390)
```

```
X = dataset[['Open','High','Low','Close','Volume']].values
y = dataset[['pred_close']].values
print(X)
print(y)
```

```
dataset
```

```
X = X[:-390]
print(X)
y = y[:-390]
print(y)
```

```
X_train = X[:len(X)-390]
X_test = X[len(X)-390:]
y_train = y[:len(y)-390]
y_test = y[len(y)-390:]
```

### Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # In deep learning it is important to feature scale all the variables even the binary ones or the dummy variables 
X_test = sc.transform(X_test) 
```

## Part 2 - Building the ANN

### Initializing the ANN

```
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

### Adding the second hidden layer

```
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
```

### Adding the output layer

```
ann.add(tf.keras.layers.Dense(units=1))
```

## Part 3 - Training the ANN

### Compiling the ANN

```
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

### Training the ANN on the Training set

```
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

## Part 4 - Making the predictions and evaluating the model

```
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
```

```
y_pred_change = np.subtract(y_pred,y2_pred)
y_real_change = np.subtract(y_test,y2_test)
```

```
np.concatenate((y_pred_change.reshape(len(y_pred_change),1),y_real_change.reshape(len(y_real_change),1)),1)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(np.sign(y_pred_change),np.sign(y_real_change)) 
```
