```
import pandas as pd
import random

limit = 1000
d = 1
data = []
while(d <= limit):
    count = 0
    Subset = False
    while(not Subset):
        a = []
        b = []

        for i in range(0,d*2):
            a.append(random.randint(0,25))

        for i in range(0,d):
            b.append(random.randint(0,25))

        Subset = True
        common_counter = 0
        common = [None]*len(b)
        for i in range(0, len(b)):
            for j in range(0, len(a)):
                if b[i]!=a[j]:
                    Subset = False
                else:
                    if common[i]!=j:
                        common_counter+=1
                        common[i] = j

                if common_counter==2:
                    Subset = True
                    break

        count+=1
        if Subset:
            break
    
    

    #print("Which is a probabilty of:",1.0/count)
    data.append(count)
    d+=1

df = pd.DataFrame(data, columns=['Counts'],index = data.index+1)
df.to_csv("Data.csv")
```
