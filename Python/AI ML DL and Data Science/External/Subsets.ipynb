```
%tensorflow_version 2.x
import tensorflow as tf
print("Tensorflow version " + tf.__version__)

try:
  tpu = tf.distribute.cluster_resolver.TPUClusterResolver()  # TPU detection
  print('Running on TPU ', tpu.cluster_spec().as_dict()['worker'])
except ValueError:
  raise BaseException('ERROR: Not connected to a TPU runtime; please see the previous cell in this notebook for instructions!')

tf.config.experimental_connect_to_cluster(tpu)
tf.tpu.experimental.initialize_tpu_system(tpu)
tpu_strategy = tf.distribute.experimental.TPUStrategy(tpu)
```

```
import pandas as pd
import random

limit = 1000
d = 1
data = []
while(d <= limit):
    count = 0
    Subset = False
    while(not Subset):
        a = []
        b = []

        for i in range(0,d*2):
            a.append(random.randint(0,25))

        print("List A:",a)

        for i in range(0,d):
            b.append(random.randint(0,25))

        print("List B:", b)

        print("Checking if B is a sublist of A..")

        Subset = True
        
        for i in a:
          if not i in b:
            Subset = False
            break

        print("B is a sublist of A" if Subset else "B is not a sublist of A")
        count+=1
        if Subset:
          print("The program ran",count,"to check if A 1x",len(b),"list is a sublist of a 1x",len(a),"list")
          break
    #print("Which is a probabilty of:",1.0/count)
    data.append(count)
    d+=1

df = pd.DataFrame(data, columns=['Counts'],index = data.index+1)
df.to_csv("Data.csv")
```
