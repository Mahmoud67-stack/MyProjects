# Predicting Housing Prices

## Importing the Libraries

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error 
from sklearn.decomposition import FastICA
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam
```

## Importing the Dataset and doing some preprocessing shit on it

```
dataset = pd.read_csv("kc_house_data.csv")
dataset
```

```
dataset.pop('id')
dataset.pop('date')
dataset.pop('zipcode')
price = dataset.price
dataset.pop('price')
dataset['price'] = price
```

```
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset
```

## Spliting the datasets into a training and test sets

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Feature Scaling to recognize the patterns in the data much better and faster


```
sc_x = MinMaxScaler(feature_range=(0, 1))
sc_y = MinMaxScaler(feature_range=(0, 1))
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

## Building the model

```
ann =  Sequential()
```

### Adding the input layer

```
ann.add(Dense(units = 64, activation = 'relu', kernel_initializer='uniform', input_dim = 17))
```

### Adding the first hidden layer

```
ann.add(Dense(units = 128, kernel_initializer='uniform', activation = 'relu'))
```

### Adding the second hidden layer

```
ann.add(Dense(units = 200, kernel_initializer='uniform', activation = 'relu'))
```

### Adding the third hidden layer

```
ann.add(Dense(units = 128, kernel_initializer='uniform', activation = 'relu'))
```

### Adding the fourth hidden layer

```
ann.add(Dense(units = 64, kernel_initializer='uniform', activation = 'relu'))
```

### Adding the output layer

```
ann.add(Dense(units = 1, kernel_initializer='uniform'))
```

## Compile the Model

```
ann.compile(optimizer = Adam(lr = 0.001), loss = 'mse', metrics = ['mean_absolute_error'])
```

```
import random
```

```
r = random.randint(0, len(X_train) - 1)
```

## Fitting the model on the data

```
ann.fit(X_train, y_train, batch_size = 32, epochs = 100, validation_data=(X_test, y_test))
```

```
ann.fit(X_train[r], y_train[r], batch_size = 1, epochs = 2, validation_data=(X_test, y_test))
```

## Prediciting the outcome 

```
y_pred = sc_y.inverse_transform(ann.predict(X_test))
```

## Comparing the results

```
print(y_pred)
```

```
y_test = sc_y.inverse_transform(y_test)
print(y_test)
```

```
print((mean_squared_error(y_test, y_pred)))
```

```
error = abs(y_pred - y_test)/y_test
print(np.mean(error))
```
