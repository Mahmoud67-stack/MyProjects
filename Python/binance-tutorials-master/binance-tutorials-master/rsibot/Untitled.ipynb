```python
import pandas as pd
import numpy as np
import talib as ta

dataset = pd.read_csv("BTCUSD.csv")
```

```python
initial = 1330.42
RSI_to_buy = 30
RSI_TO_sell = 70
```

```python
dataset['RSI'] = ta.RSI(dataset['Close'],timeperiod=14)
```

```python
dataset
```

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
RSI_Data = np.array(dataset['RSI'])
Close = np.array(dataset['Close'])
```

```python
bought_price = 0
sold_price = 0
profit = 0
bought = False
sold = False
total_balance = initial
for i in range(len(RSI_Data)):
    if(RSI_Data[i] <= 20 and not bought):
        print("Buying at", Close[i])
        bought_price = Close[i]
        bought = True
        sold = False
    elif(RSI_Data[i]>=80 and not sold):
        print("Selling at", Close[i])
        sold_price = Close[i]
        profit = (sold_price-bought_price)/bought_price
        total_balance = total_balance*(1+profit)
        print("Your balance now =", total_balance)
        sold = True
        bought = False
        
```
