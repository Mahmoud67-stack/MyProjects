```python
import pandas as pd
from talib import abstract
import talib
import numpy as np
df  = pd.read_csv("BTCUSDT-1m-data.csv")

list_of_functions = talib.get_functions()
df
```

```python
data = []
for function in list_of_functions:
    if function=='MAVP':
        pass
    else:
        # or by name
        funcname = abstract.Function(function)
        myString = "abstract."+str(function)+"(df, timeperiod = 14)"
        my_list = eval(myString)
        df1 = pd.DataFrame(my_list)
        if df1.columns.any()==0:
            df1.rename(columns = {0:function}, inplace = True)
        df = pd.concat([df, df1], join = 'outer', axis = 1)
        print(df)
df.to_csv("BTC_Modf_lib.csv")
```
